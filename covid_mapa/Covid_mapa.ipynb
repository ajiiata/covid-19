```python
merged = pd.merge(df1, countries)
```

```python
merged1 = merged.drop(['Date_reported'], axis = 1)
merged2 = merged1.groupby(by = ['COUNTRY']).agg(Cumulative_cases = ('Cumulative_cases', 'max'), geometry = ('geometry', 'first'), cumulative_deaths = ('Cumulative_deaths', 'max'))
```

```python
merged3 = merged2.append(tmp_df)

gdf = gpd.GeoDataFrame(merged3, geometry='geometry')

gdf1 = gdf.sort_values(by='cumulative_deaths', ascending=True, na_position='first')
```

```python
plt.style.use('default')
```

```python
def truncate_colormap(cmap, minval=0.0, maxval=1.0, n=1000):
    new_cmap = colors.LinearSegmentedColormap.from_list('trunc({n},{a:.2f},{b:.2f})'.format(n=cmap.name, a=minval, b=maxval),cmap(np.linspace(minval, maxval, n)))
    return new_cmap

arr = np.linspace(0, 50, 100).reshape((10, 10))
fig, ax = plt.subplots(ncols=2)
cmap = plt.get_cmap('RdPu')
cmap2 = plt.get_cmap('PuRd')
new_cmap2 = truncate_colormap(cmap2, 0.3, 1)
new_cmap = truncate_colormap(cmap, 0.3, 1)

# normalize color cumulative cases:

def ifnan(number):
    return not math.isnan(number)

a = gdf1['Cumulative_cases'].values.tolist()
tmp = list(filter(ifnan , a))

vmin, vmax, vcenter = min(tmp), max(tmp), 10000000
norm = TwoSlopeNorm(vmin=vmin, vcenter=vcenter, vmax=vmax)


# create a normalized colorbar
cbar = plt.cm.ScalarMappable(norm=norm, cmap=new_cmap)

#--------------------------------------------------------------------------------------------------------------------------

b = gdf1['cumulative_deaths'].values.tolist()
tmp1 = list(filter(ifnan , b))

vmin1, vmax1, vcenter1 = min(tmp1), max(tmp1), 40000
norm1 = TwoSlopeNorm(vmin=vmin1, vcenter=vcenter1, vmax=vmax1)
# create a normalized colorbar
cbar1 = plt.cm.ScalarMappable(norm=norm1, cmap=new_cmap2)
```

```python
plt.style.use('dark_background')
```

```python
font = {'family' : 'DejaVu Sans'}
plt.rc('font', **font)

fig, ax = plt.subplots(1, figsize=(20, 10))
ax.axis('off')
ax.set_title('Zachorowania na COVID-19', fontdict={'fontsize': '20', 'fontweight' : 'bold'})
im = gdf.plot(column='Cumulative_cases',
            cmap=new_cmap,
            norm=norm,
            linewidth=0.9,
            ax=ax,
            edgecolor='0',
            legend=False, 
         legend_kwds={'label': "Liczba zarażeń [w milionach]",
                      'orientation': "vertical",
                       'pad': 0.01},
             missing_kwds={
            "color": "lightgrey",
            "label": "Missing values",},)

frame1 = plt.gca()
frame1.axes.get_xaxis().set_visible(False)
frame1.axes.get_yaxis().set_visible(False)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

fig.colorbar(cbar, ax=ax)
ax.grid(False)

# plt.savefig('covid_cases_choropleth_map.png',dpi = 800, transparent = True)
plt.show()
```
