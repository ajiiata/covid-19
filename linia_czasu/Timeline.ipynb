## Timeline pokazujący, kiedy (mniej więcej*) wydarzyła się jaka pandemia

#### *mniej więcej - nie zawsze dysponujemy idealnymi danymi, które mówią co do dnia, kiedy coś się wydarzyło. 

-------------------------------------------------------------------------------------------------------------------------------

Co w sytuacji, kiedy nie mamy:
- *dnia* - wtedy zakładamy, że coś stało się 15-go danego miesiąca
- *miesiąca* - zakładamy, że coś stało się 6-go miesiąca
- *roku* - zakładamy, że coś stało się w połowie wieku, chyba że jest napisane np. na początku, to damy 1/5-tą wieku, pod koniec damy 4/5-te wieku 

### Import pakietów:

```python
import matplotlib.pyplot as plt
import matplotlib.pylab as plt
```

### Zmiana tła wykresu na czarne:

```python
plt.style.use('dark_background')
```

```python
def timeline(event_title, event_year, event_text, event_delta_y, file_name):
    scale = 10
    margin = 1
    margin_delta = 1.0 / (scale + margin * 2) 
    
    min_date = min(event_year)
    max_date = max(event_year)
    date_range = max_date - min_date
    date_pos = scale / date_range
    
    date_x = [(d - min_date) * date_pos + margin for d in event_year]
    date_y = [0 for d in event_year]
    line_y = [1 for d in event_year]
    
    fig, ax = plt.subplots(figsize=(20, 10),  constrained_layout=True)
    ax.set_ylim(-2, 2)
    ax.set_xlim(0, scale + margin * 2)
    ax.set_title(event_title, fontweight="bold", fontfamily='Verdana', fontsize=35, color='palegoldenrod')
    
    ax.axhline(0, xmin = margin_delta, xmax = 1-margin_delta, c='palevioletred', zorder=1)
    
    ax.scatter(date_x, date_y, s=140, c='darkmagenta', zorder=2)
    ax.scatter(date_x, date_y, s=50, c='mediumorchid', zorder=3)
    
    fsize = 21.5
    delta_text = 0.1
    
    for x, d, text, delta_y  in zip(date_x, event_year, event_text, event_delta_y ):


        delta_text_y = delta_y + delta_text if delta_y > 0 else delta_y - delta_text
        v = 'bottom' if delta_y > 0 else 'top'
        
        props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)

        ax.text(
            x, delta_text_y, str(d)+", " + text,
            ha='center', va=v, 
            fontfamily='Verdana', fontweight='bold', fontsize=fsize,
            color='indigo', bbox=props
        )
        
    markerline, stemline, baseline = ax.stem(date_x, event_delta_y, use_line_collection=True)

 
    plt.setp(baseline, zorder=0)
    plt.setp(markerline, marker='', color='#4a4a4a')
    plt.setp(stemline, color='peru')


    ax.set_xticks([]) 
    ax.set_yticks([]) 


    for spine in ["left", "top", "right", "bottom"]:
        ax.spines[spine].set_visible(False)

    file_name = str(file_name + ".png")
#     plt.savefig(file_name, dpi = 800, transparent = True)
    plt.show()
    
```

```python
title = "Data rozpoczęcia danej pandemii"


year = [165, 541, 1346, 1852, 1889, 1899, 1918, 1957, 1968, 2005, 2019]   


event_text = ["Zaraza Antoninów",
                  "Dżuma Justyniana", 
                  "Czarna śmierć (The Black Death)",
                  "Trzecia pandemia cholery", 
                  "Grypa",
                  "Szósta pandemia cholery",
                  "Pandemia grypy hiszpanki",
                  "Grypa azjatycka",
                  "Grypa z Hongkongu",
                  "Pandemia HIV/AIDS", 
                  "Pandemia COVID-19"]

event_delta_y = [0.5, -0.5, 0.5, -0.5, 0.5, -1, 1, -1.5, 1.5, -2, 2]
event_delta_y = [0.85*i for i in event_delta_y]

timeline(title, year, event_text, event_delta_y, "timeline1")
```

```python
title = "Data rozpoczęcia danej pandemii\n - czasy współczesne"


year = [1852, 1889, 1899, 1918, 1957, 1968, 2005, 2019]   


event_text = [    "Trzecia pandemia cholery", 
                  "Grypa",
                  "Szósta pandemia cholery",
                  "Pandemia grypy hiszpanki",
                  "Grypa azjatycka",
                  "Grypa z Hongkongu",
                  "Pandemia HIV/AIDS", 
                  "Pandemia COVID-19"]

event_delta_y = [0.5, -0.5, 1, -1, 1.5, -1.5, 2, -2]
event_delta_y = [0.75*i for i in event_delta_y]

timeline(title, year, event_text, event_delta_y, "timeline2")
```
